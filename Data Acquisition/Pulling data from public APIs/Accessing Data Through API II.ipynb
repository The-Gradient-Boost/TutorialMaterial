## Part 1: Concepts

#### 1. API

API is an acronym for Application Programming Interface.

A simple everyday analogy that mirrors the role of an API is the restaurant waiter/ess.

An API is simply an interface that allows you to communicate with a server. There are a lot of communications that can be done between you as a client and the server. These set of communications are known as Methods.

##### Why APIs

API, as earlier mentioned, helps us to communicate with servers. There are a number of reasons why an organization may want to create an interface for us to communicate with her servers. A number of them include: 
<ul>
    <li> To grant access to the general public where data should be made public
</ul>
    
<ul>
    <li> To somehow lure you into buying a premium plan should you want to access more data frequently
</ul>
    
<ul>
    <li> To reduce the chances of crashing their server due to overload from requests since they know you could be accesing the data by scrapping from the websites
</ul>
<ul>
    <li> etc.
</ul>

##### Where To Find an API

It is often the case that organizations provide an API link on thier website. In other cases, one may have to search the internet for popular APIs to the platform they want to acces.

You can find a list of some popular public APIs [here](https://github.com/public-apis/public-apis)

However, don't be distracted as that is not the goal of this tutorial. The link is for you to have a resource, not tp be carried away by it

#### 2. Structure of an API

APIs are just like the web urls you are used to seeing. The oly difference is that they can not be used as one, but as was intended; as arguments in an API wrapper or requests libraries

Each url is a request and the data sent back is a response. Most APIs return data in JSON formats. You should be aware of this.

![alt_text](https://phpenthusiast.com/theme/assets/images/blog/what_is_rest_api.png?021019a)

Below is an example of an API, this is the github api

https://api.github.com

##### a. Endpoints

This is the url you are requesting for. It is the starting point of the API you are requesting from.
Like the Github API above, what is shown is the root-endpoint

##### b. Methods

The method is the type of request you send to the server. It gives meaning to the endpoint and the request you are making. There are up to five methods one can apply.

<ul>
    <li> GET : This is used for retrieving data from the server. This is mostly what we will be doing in this tutorial as we seek to pull data from platforms
</ul>
    
<ul>
    <li> POST : This is used to create new data in the database
</ul>
<ul>
    <li> PUT : This is used to update data on the database
</ul>
    
<ul>
    <li> PATCH : This is used to update data on the database
</ul>
<ul>
    <li> DELETE : This is used to delete data from a database
</ul>
 

##### c. Headers

Headers are used to provide information to both the client and server. It can be used for many purposes, such as authentication and providing information about the body content.

##### d. Data

The data (sometimes called “body” or “message”) contains information you want to be sent to the server. This option is only used with POST, PUT, PATCH or DELETE requests.

#### 3. Working With APIs

APIs can be documented well, poorly, or somewhere in between. If you’re lucky, they are well-documented. If you’re really lucky there is a Python wrapper for the API that works and is well documented.

##### API Wrappers

An API wrapper is simply a python library that is custome developed for accessing a particular API. Just like other python libraries, they have packages in them that allows users perform a variety of operations as suited for the platfrom in question.

As mentioned earlier, one can be lucky to find an API wrapper that exists for their intended purpose by searching on the official website of the organization or by using Google.

I have provided a list of API Wrappers curated and maintained by Jeff Hale on his Github repo [here](https://github.com/discdiver/list-of-python-api-wrappers)

##### Requests

Now we know that we can make use of API tools such as wrappers to access an API terminal. However, you won't always have an already made custom tool, hence you need to find an alternative to requesting data from an API.

This is where the Requests library comes in.

Requests is a Python library for accessing web data. As at today, it is the most downloaded Python package.

We will see how this is used to access data from an API in the Practice session.

#### 4. API Documentation

An API documentation is simply a guide or user manual on what type of resources can be accessed through the API, how to access them (what methods to call or parameters to use) and tells you of the do(s) and don'ts of using the API.

Typically, the documentation will show you a sample response for any of the availabel methods. That is, it will tell what type of data is returned e.g, JSON, and what a typical format looks like. This is so you can be sure of what you need and how to retrieve them.

An example of an API documentation is found here, [Coinbase](https://developers.coinbase.com/api/v2) and [last fm](https://www.last.fm/api/intro)

As can be seen on both sample documentations, the rules of engagement are spelt out. Requirements for general use and for method use are also spelt out. I advise you stick to the rules outlined by the API provider to avoid being banned from using the service.

##### Documentation Notes

Some other things you should know before we proceed to the practice session are:

<ul>
    <li> API Keys: An API key is sometimes required when you need to perform some special operations on user accounts. An example will be sending a Bitcoin on the Coinbase platform or deleting a deleting a playlist on the Lastfm platform. To get an API Key, some providers will require you to create an account with them by signing up.
        
        It is important to mention that your API Keys are very delicate and as such, should be kept safe from the public. Do not save them in public repos or write them down carelessly as they can be used to access private data on the platform.
</ul>
<ul>
    <li> Rate Limits: Usually, the API provider will limit the amount of calls/requests made on the server per second. This is to avoid causing problems on the server due to multiple requests and overloading. You should restrict the rate of calls/request to avoind being banned from accessing the service.
</ul>

Having understood how to retrieve price data (spot_price), let's now retrieve define a function that allows us to retrieve data for date ranges.

```python
# Import date pakages

from datetime import date, timedelta, datetime #for date and time handling
```

```python
# Write a function for retrieving historical spot_prices

def price_history(startdate, enddate):
    '''This retrieves historical spot_price data of the Bitcoin in USD from the Coinbase API. Arguments should be string types'''
    
    # create an empty list for storing price data
    prices=[]
    
    #Use strptime method of datetime to parse string inputs dates
    #Note that we have declared a format for the dates to be entered, example, '28-02-2020' and not '28 Feb 2020', i.e 'DD-MM-YYY'
    d1 = datetime.strptime(startdate, '%d-%m-%Y')
    d2 = datetime.strptime(enddate, '%d-%m-%Y')
    
    #Use list comprehension to generate a range of date from start to end of history
    date_range = [d1 + timedelta(days=x) for x in range((d2-d1).days + 1)]
    
    #Loop over the dates in the date_range and scoop spot_prices for each day
    for day in date_range:
        spot_price = client.get_spot_price(currency_pair = 'BTC-USD', date=day)
        
        #Add the date to the returned response. Note that responses are in JSON format (python dictionary)
        spot_price['Date'] = day
        
        #Finally, append the spot_price to the list of prices 
        prices.append(spot_price)
        
    #The functions return the list of historical prices
    return prices
```

The above output is a JSON string. We need to convert this to a python object (dictionary). We make use of the JSON dumps method of JSON

```python
import json
```

```python
# Create a function that prints out the JSON formatted string

def print_json(response):
    word = json.dumps(response, sort_keys=True, indent=4)
    
    print(word)

    
print_json(req.json())  # this encodes the string into JSON formatted string
```

The above is a JSON formatted response. Let's take a sneak peek at the @attr key

```python
#Print the @attr key
print_json(req.json()['tracks']['@attr'])
```

There are a total of 27,092,264 tracks and 541,846 pages of data. By default, the number of data fetched per page is 50. Data returned earlier was just for the first page.

We will go ahead and attempt to write a function to fetch data from the first 200 pages (for the purpose of this tutorial. You can fetch more when you have more time, personally). We will also increase the limit to fetch more data per page to improve the number of data we're getting since we reduced the page size due to rate considerations.

Talking about rate considerations, we don't want to be banned from using the API as a result of repeated calls within a short time, hence we will delay each call using a sleep method from the time package in Python.

```python
#A function to achieve the above and any other endpoint method is thus

def lastfm_data(endpoint):
    
    # define headers and URL
    header = {'user-agent': user}
    link = 'http://ws.audioscrobbler.com/2.0/'

    # Add API key and format to the load
    endpoint['api_key'] = api_key
    endpoint['format'] = 'json'

    response = requests.get(link, headers=header, params=endpoint)
    return response
```

```python
lastfm_data({'method':'chart.gettopartists'})
```

```python
#Import time package
import time

#To clear out the output in order to not have a messy notebook
from IPython.core.display import clear_output
```

```python
#Create empty list for appending responses
responses = []

page = 1
total_pages = 200 # this is number of pages we want to fetch data from

while page <= total_pages:
    load = {
        'method': 'chart.gettoptracks',
        'limit': 500,
        'page': page
    }

    # print some output so we can see the status of the operation
    print("Requesting page {}/{}".format(page, total_pages))
    
    # clear the output to make things neater
    clear_output(wait = True)

    # make the API call
    response = lastfm_data(load)

    # if we get an error, print the response and halt the loop
    if response.status_code != 200:
        print(response.text)
        break

    # extract pagination info
    page = int(response.json()['tracks']['@attr']['page'])
    
    # append response
    responses.append(response)

    #Apply a sleep of 10% of a second 
    time.sleep(0.1)

    # increment the page number
    page += 1
```

#### Dealing with JSON data

As earlier mentioned and as seen above, most API return data in JSON formats. This is the primary language of APIs. We will see how we can handle this data and retrieve the neccesary we are interested in.

When we receive a response from the API, the first thing to do is to call a .json() method on the response in order to see the retrieved data.

```python
#Let us take the first response from our list of responses

res1 = responses[0]

#Call a .json() method on it to see the data
res1_json = res1.json()

#printing this out reveals the data
print(res1_json)
```

However, this does not allow us to make much sense of the data as it is not formatted in JSON style. Python has a good JSON support with the json package. 


Two important methods exists viz, json.dumps() and json.loads(). The first takes a Python object and converts it to a string while the second takes a JSON string and converts it to a Python object.

You can use of the first to format responses to JSON string formats for readability.

```python

res1_tracks = res1_json['tracks']['track']
#print(res1_tracks)
res1_df = pd.DataFrame(res1_tracks)

res1_df.head()
```
